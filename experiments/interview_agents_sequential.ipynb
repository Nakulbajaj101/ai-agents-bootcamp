# Building the AI Interviewer Agent

```python
import sys

sys.path.insert(0, '..')
from crewai.agent import Agent
from crewai.process import Process
from crewai.crew import Crew
from crewai.task import Task
from crewai.knowledge.source.string_knowledge_source import StringKnowledgeSource
from config import JOB_DESCRIPTION, JOB_TITLE, COMPANY_LOCATION, COMPANY_NAME, COMPANY_HISTORY, HIRING_MANAGER_EXPERIENCE
```

```python
from IPython.display import display, Markdown
```

```python
interview_agent_role=f"""You are an interviewer agent simulator,
    who is interviewing for the position of {JOB_TITLE} for company {COMPANY_NAME}
    at location {COMPANY_LOCATION}"""
interview_agent_goal=f"""You will help the user to prepare for the interview for job position 
    {JOB_TITLE} with job description {JOB_DESCRIPTION}
    by asking questions related and relevant for the position"""
```

```python
interviewer_agent = Agent(
    role=f"{interview_agent_role}",
    goal=f"{interview_agent_goal}",
    backstory=f"""Company history is : 
    <COMPANY HISTORY>
    {COMPANY_HISTORY}
    <COMPANY HISTORY>
    \n
    and interviewer is David Tan whose bio is below:
    {HIRING_MANAGER_EXPERIENCE}
    """,
    llm="gpt-4o-mini"
)
```

```python
interviewer_prep_task = Task(
    name="interview_questions",
    description=f"Asking the user questions related to the {JOB_TITLE} with following descption: {JOB_DESCRIPTION}",
    agent=interviewer_agent,
    expected_output=f"Asking the user one question at a time related to the job title {JOB_TITLE}",
    human_input=True
)
```

```python
# AI Crew
crew = Crew(
    name="Interview_process_crew",
    agents=[interviewer_agent],
    tasks=[interviewer_prep_task],
    verbose=True,
    process=Process.sequential
)
```

```python
result  = crew.kickoff()
```

# Building the Interviewer Coach 

```python
interviewer_coach = Agent(
    role="Interview Coach",
    goal=f"""Help the user/interviewee prepare for the job interview for position {JOB_TITLE} 
    by grading the relevance and quality of answers to the interview questions, and providing feedback
    to improve""",
    backstory="""You are an expert in interview preparation coaching and mentoring 
    for job positions in Australia for senior, lead, principal and managerial Machine learning Engineer
    positions that focus on software and ml design best practices as well as team dynamics in 
    ML cross functional teams""",
    llm="gpt-4o-mini"
)

interviewer_coach_task = Task(
    name="interview_feedback",
    description=f"Giving feedback and grading associated with answers for intervier for job position {JOB_TITLE}",
    agent=interviewer_coach,
    expected_output="""Give feedback to the user in bullet points, 
                    and grading the answer between 1 and 5, 
                    and help them improve by making feedback constructive""",
    context=[interviewer_prep_task]
)
```

```python
# AI Crew
crew = Crew(
    name="Interview_process_crew",
    agents=[interviewer_agent, interviewer_coach],
    tasks=[interviewer_prep_task, interviewer_coach_task],
    verbose=True,
    process=Process.sequential
)
```

```python
result = crew.kickoff()
```

```python
for res in result.tasks_output:
    display(Markdown(res.raw))
```

# Lets add memory and knowledge

```python
interview_crew = Crew(
    name="Interview_question_crew",
    agents=[interviewer_agent],
    tasks=[interviewer_prep_task],
    verbose=True,
    process=Process.sequential
)

interview_questions = []
result = interview_crew.kickoff()
for res in result.tasks_output:
    interview_questions.append(res.raw)
interview_content = "\n".join(interview_questions)
string_source = StringKnowledgeSource(content=interview_content)


interviewer_coach = Agent(
    role="Interview Coach",
    goal=f"""Help the user/interviewee prepare for the job interview for position {JOB_TITLE} 
    by grading the relevance and quality of answers to the interview questions, and providing feedback
    to improve""",
    backstory="""You are an expert in interview preparation coaching and mentoring 
    for job positions in Australia for senior, lead, principal and managerial Machine learning Engineer
    positions that focus on software and ml design best practices as well as team dynamics in 
    ML cross functional teams""",
    llm="gpt-4o-mini"
)

interviewer_coach_task = Task(
    name="interview_feedback",
    description=f"Giving feedback and grading associated with answers for intervier for job position {JOB_TITLE}",
    agent=interviewer_coach,
    expected_output="""Give feedback to the user in bullet points, 
                    and grading the answer between 1 and 5, 
                    and help them improve by making feedback constructive""",
    context=[interviewer_prep_task]
)

interviewer_coach_crew = Crew(
    name="Interview_feedback_crew",
    agents=[interviewer_coach],
    tasks=[interviewer_coach_task],
    process=Process.sequential,
    memory=True,
    knowledge_sources=[string_source]
)

feedback_result = interviewer_coach_crew.kickoff()
for res in feedback_result.tasks_output:
    display(Markdown(res.raw))
```
